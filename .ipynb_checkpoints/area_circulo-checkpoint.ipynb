```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
n = 1000
randx = np.random.randn(n)
randy = np.random.randn(n)

a = 2
r = a / 2
k = 10

plt.figure(figsize=(11,11))

ax1 = plt.subplot(221)

c1 = plt.Circle(xy=[0,0], radius=r, alpha=0.4, color='red')
ax1.add_artist(c1)
plt.xlim(-1,1)
plt.ylim(-1,1)

ax2 = plt.subplot(222)

for xy in xys:
    c2 = plt.Circle(xy = xy, radius=r2, alpha=0.4, color='red')
    ax2.add_artist(c2)
plt.xlim(-1,1)
plt.ylim(-1,1)

###### GRAFICO MONTECARLO ######

ax3 = plt.subplot(223)

plt.scatter(randx, randy, alpha=0.6, color='black')

c1 = plt.Circle(xy=[0,0], radius=r, alpha=0.4, color='red')
ax3.add_artist(c1)
plt.xlim(-1,1)
plt.ylim(-1,1)

ax4 = plt.subplot(224)

origen = -1
a2 = a / k
r2 = a2 / 2
x = np.arange(k) * a2 + r2 - 1
y = np.arange(k) * a2 + r2 - 1 
xys = [[xi,yi] for xi in x for yi in y]

plt.scatter(randx, randy, alpha=0.7, color='black')
count = 0
for xy in xys:
    c2 = plt.Circle(xy = xy, radius=r2, alpha=0.4, color='red')
    ax4.add_artist(c2)
plt.xlim(-1,1)
plt.ylim(-1,1)
```

```python
x1 = np.arange(k) * a2
x2 = x1 + a2
areas = list(zip(x1,x2))
count2 = 0

for ax in areas:
    for ay in areas:
        for i in range(n):
            x = randx[i]
            y = randy[i]
            cond1 = ax[0] <= x <= ax[1]
            cond2 = ay[0] <= y <= ay[1]
            if cond1 and cond2:
                if (x ** 2 + y ** 2) ** 0.5 > r2:
                    count2 += 1

count1 = 0
for i in range(n):
    x = randx[i]
    y = randy[i]
    if (x ** 2 + y ** 2) ** 0.5 > r:
        count1 += 1
```

```python
plt.barh(0.2, count1 / n, 0.1)
plt.barh(0.4, count2 / n, 0.1)
plt.text(count1 / n - 0.2, 0.25, 'K=1, Area=' + str(count1/n), color='white')
plt.text(count2 / n - 0.21, 0.45, 'K=100, Area=' + str(count2/n), color='white')
```
